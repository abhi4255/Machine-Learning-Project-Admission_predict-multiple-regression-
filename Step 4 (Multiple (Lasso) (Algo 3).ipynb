```python
x=df[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA',
       'Research']]
y=df['Chance of Admit ']
```

# Train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=8)
```

```python
x_train
```

```python
x_test
```

# Step-4,5 : Modelling & Evalution

## Lasso

**Applying Hyperparameter tunning for Lasso Regression**

```python
from sklearn.model_selection import GridSearchCV

# model
from sklearn.linear_model import Lasso
estimator=Lasso()

# Parameters & Values
param_grid = {"alpha":list(range(1,100))}

# Identifying the best values of the parameter within given values fo the given data
model_hp = GridSearchCV(estimator,param_grid,cv=5, scoring="r2")

model_hp.fit(x_train,y_train)

model_hp.best_params_
```

# Build a Lasso Model using the best hyperparameters

```python
# Modeling

from sklearn.linear_model import Lasso
lasso_best = Lasso(alpha=1)
lasso_best.fit(x_train,y_train)

print("Intercept:",lasso_best.intercept_)
print("coefficents:",lasso_best.coef_)

# predication & Evaluation on train data
ypred_train = lasso_best.predict(x_train)

from sklearn.metrics import r2_score  
# Here train R2 and algo 1 R2 simple linear are the same because loss is not working, Lasso and what are the penalty term is almost 0
print("train R2:",r2_score(y_train,ypred_train))

# Cross Validation Score
from sklearn.model_selection import cross_val_score
print("CV Score:",cross_val_score(lasso_best,x_train,y_train,cv=5,scoring="r2").mean())

# Prediction on test data
ypred_test = lasso_best.predict(x_test)

# Evalution on test data
print("test R2:",r2_score(y_test,ypred_test))
```

# Final Lasso Model
* with best hyperparameter & with important feature

```python
df.drop(columns=['TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA',
       'Research'],inplace=True)

x=df.drop('Chance of Admit ',axis=1)
y=df['Chance of Admit ']


x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=8)

# Modeling

from sklearn.linear_model import Lasso
lasso_best = Lasso(alpha=1)
lasso_best.fit(x_train,y_train)

print("Intercept:",lasso_best.intercept_)
print("coefficents:",lasso_best.coef_)

# predication & Evaluation on train data
ypred_train = lasso_best.predict(x_train)

from sklearn.metrics import r2_score  
# Here train R2 and algo 1 R2 simple linear are the same because loss is not working, Lasso and what are the penalty term is almost 0
print("train R2:",r2_score(y_train,ypred_train))

# Cross Validation Score
from sklearn.model_selection import cross_val_score
print("CV Score:",cross_val_score(lasso_best,x_train,y_train,cv=5,scoring="r2").mean())

# Prediction on test data
ypred_test = lasso_best.predict(x_test)

# Evalution on test data
print("test R2:",r2_score(y_test,ypred_test))
```
