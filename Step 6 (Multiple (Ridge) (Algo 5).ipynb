# Apply Ridge Algorithm

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter("ignore")
```

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

from sklearn.model_selection import GridSearchCV
```

```python
df=pd.read_excel(r"C:\Users\abhis\DATA SCIENCE SUBJECT & AI\Machine Learning\PROJECTS\Admission_predict ( Multiple Regression )(MR) Project-(3)\cleaned_data.xlsx")
df.head()
```

```python
df=df.drop(columns=['Unnamed: 0'])
df
```

```python
df.columns
```

```python
x=df.drop('Chance of Admit ',axis=1)
y=df['Chance of Admit ']
```

```python
df.shape
```

# Train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=8)
```

# Step-4,5 : Modelling & Evalution

**Applying Hyperparameter tunning for Ridge Regression**

```python
# find the alpha value

from sklearn.model_selection import GridSearchCV

estimator = Ridge()

# When I try to increase every time, the last number gets the best number. So indirectly is saying that Ridge is not working.
param_grid = {"alpha":list(range(1,100))}

model_hp= GridSearchCV(estimator,param_grid,scoring="r2", cv=5)  

model_hp.fit(x_train,y_train)

model_hp.best_params_


# put the alpha value in the ridge code
```

## Modeling Ridge Regression using best hyperparameters Code

```python
# Modeling

ridge_best = Ridge(alpha=1)
ridge_best.fit(x_train,y_train)

print("Intercept:",ridge_best.intercept_)
print("Coefficients:",ridge_best.coef_)
```

# Step 5 Evaluation

##  Evaluation on Train Data

```python
%%timeit
ypred_train=ridge_best.predict(x_train)
print("Train R2:",r2_score(y_train,ypred_train))
print("CV Score:",cross_val_score(ridge_best,x_train,y_train,cv=5).mean())
```

## Evaluation of Test Data

```python
ypred_test = ridge_best.predict(x_test)
print("Tets R2:",r2_score(y_test,ypred_test))
```
